### Step 1: Import libraries and data

```python
# Step 1: Import libraries and data
import pandas as pd 
import numpy as np 

df = pd.read_csv("C:\\Users\\chase\\Downloads\\Preppin' Summer 2022 - PD 2022 Wk 27 Input.csv")
df
```

### Step 2: Separate out the Product Name field to form Product Type and Quantity

```python
fields = df["Product Name"].str.split(pat = ' - ', expand = True)
df[['Product Type', 'Quantity']] = fields
df
```

### Step 3: Create two paths in your flow: (1) One to deal with the data about Liquid Soap sales, (2) One to deal with the data about Bar Soap sales

```python
liquid_df = df.loc[df['Product Type'] == "Liquid", :]
bar_df = df.loc[df['Product Type'] == "Bar", :]
print("Liquid DataFrame")
display(liquid_df)
print("Bar DataFrame")
display(bar_df)
```

### Step 4: For each dataframe, (1) Clean the Quantity field to just leave numeric values (for liquid, ensure every value is in millilitres), (2) Sum up the sales for each combination of Store, Region, and Quantity (Name this field "Present in N Orders")

```python
liquid_df.loc[liquid_df['Quantity'] == "1L", "Quantity"] = "1000ml"
liquid_df['Quantity'] = liquid_df['Quantity'].str.replace('ml', "")

bar_df['Quantity'] = bar_df['Quantity'].str.replace('x', "")

final = pd.DataFrame(liquid_df.groupby(['Store Name', 'Region', 'Quantity'])['Sale Value'].sum()).reset_index()
semi_final = pd.DataFrame(liquid_df.groupby(['Store Name', 'Region', 'Quantity'])['Order ID'].nunique()).reset_index()
final['Present in N Orders'] = semi_final['Order ID']
final = final.style.format({'Sale Value': '{:,.2f}'})
```
