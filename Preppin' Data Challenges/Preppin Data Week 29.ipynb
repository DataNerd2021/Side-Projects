```python
import pandas as pd
import numpy as np
```

```python
# Import data
sales = pd.read_csv("C:\\Users\\Chase\\Downloads\\Preppin' Summer 2022 - PD 2022 Wk 27 Input.csv")
targets = pd.read_csv("C:\\Users\\Chase\\Downloads\\Preppin' Summer 2022 - Targets (k's).csv")
sales['Product Name'].unique()
```

```python
# Remove unnecessary values from the Product Name field to just leave the Product Type
sales['Product Name'] = np.where(sales['Product Name'].isin(['Liquid - 25ml', 'Liquid - 50ml', 'Liquid - 100ml', 'Liquid - 250ml', 'Liquid - 500ml', 'Liquid - 750ml', 'Liquid - 1L']), 'Liquid', sales['Product Name'])
sales['Product Name'] = np.where(sales['Product Name'].isin(['Bar - 1x', 'Bar - 2x', 'Bar - 4x']), 'Bar', sales['Product Name'])
sales
```

```python
# Total Sales for each Store and Product Type
sales = sales.groupby(['Store Name', 'Region', 'Product Name'])['Sale Value'].sum().reset_index()
sales = pd.DataFrame(sales)
sales
```

```python
# Change the Targets data set into three columns (Product, Store, Sales, and  Target (k's))
targets = targets.melt(id_vars='PRODUCT', value_vars=['CHELSEA', 'DULWICH', 'LEWISHAM', 'NOTTING HILL', 'SHOREDITCH', 'WIMBLEDON'], var_name='Store', value_name='Target')
targets
```

```python
# Multiply the Target column by 1000 to create the full sales target number
targets['Target'] = targets['Target'] * 1000
targets
```

```python
# Prepare your data sets for joining by changing targets dataframe values to title case
targets['PRODUCT'] = targets['PRODUCT'].str.title()
targets['Store'] = targets['Store'].str.title()
targets.rename(columns = {'PRODUCT': 'Product'}, inplace = True)
targets
```

```python
# Join the data sets together and remove any duplicated fields
final = sales.merge(targets, how='inner', left_on='Store Name', right_on='Store')
final = final.drop(columns = ['Product Name', 'Store Name'])
final
```

```python
# Calculate whether each product in each store beats the target
final['Beats Target?'] = np.where(final['Sale Value'] >= final['Target'], 'True', 'False')
final = final[['Beats Target?', 'Target', 'Store', 'Region', 'Sale Value', 'Product']]
final = final.drop_duplicates(subset=['Region', 'Store', 'Product'])
final
```
