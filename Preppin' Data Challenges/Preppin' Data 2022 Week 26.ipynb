```python
# import modules
import pandas as pd
import numpy as np
```

```python
# input and preview the data
data = pd.read_csv("C:\\Users\\Chase\\OneDrive\\Documents\\Pandas Practice\\Sources\\Spotify Data Unclean.csv")
data.head(10)
```

```python
# Create a new field which would break down milliseconds into seconds and minutes
data['mins_played'] = (data['ms_played']/1000)/60
data.head(10)
```

```python
# Extract the year from the timestamp field
data['ts'] = pd.to_datetime(data['ts'])
data['year'] = data['ts'].dt.year
data.head(10)
```

```python
# Rank the artist by total minutes played overall
data2 = data.groupby(['Artist Name'])['mins_played'].sum()
data2 = pd.DataFrame(data2)
data2 = data2.reset_index()
data2['Overall Rank'] = data2['mins_played'].rank(method='dense', ascending=False)
data2['Overall Rank'] = data2['Overall Rank'].astype(int)
data2.sort_values(by=['Overall Rank'], ascending=True).head(10)
```

```python
# For each year, find the ranking of the artist by total minutes played
data3 = data.groupby(['Artist Name', 'year'])['mins_played'].sum()
data3 = pd.DataFrame(data3)
data3 = data3.reset_index()
data3['Rank'] = data3.groupby(['year'])['mins_played'].rank(method='dense', ascending=False)
data3['Rank'] = data3['Rank'].astype(int)
data3.sort_values(by=['Rank', 'year'], ascending=[True, True]).head(10)
```

```python
# Reshape the data so we can compare how artist position changes year to year
df = data2.merge(data3, left_on='Artist Name', right_on='Artist Name')
df = df[['Overall Rank', 'Artist Name', 'year', 'Rank']]
df['Overall Rank'] = df['Overall Rank'].astype(int)
df['Rank'] = df['Rank'].astype(int)
df = df.pivot_table(index=['Overall Rank', 'Artist Name'], columns='year', values='Rank', fill_value=0, aggfunc='sum')
df = df.reset_index()
df
```

```python
# Filter to the overall top 100 artists
df = df.loc[df['Overall Rank'] <= 100, :]
df = pd.DataFrame(df)
df
```
