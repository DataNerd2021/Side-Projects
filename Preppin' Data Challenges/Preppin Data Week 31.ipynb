```python
# Import modules
import pandas as pd
import numpy as np
import tkinter as tk
import easygui as g 
from easygui import *
import sys
```

```python
# Input the data
data = pd.read_csv("C:\\Users\\Chase\\Downloads\\Preppin' Summer 2022 - PD 2022 Wk 27 Input.csv")
data
```

```python
# Split the Product Name field into Prioduct Type and Size
data[['Product Type', 'Size']] = data['Product Name'].str.split(' - ', n=2, expand=True)

# Drop the Product Name Column
data = data.drop(columns=['Product Name'])

# Only keep the Liquid Products
data = data.loc[data['Product Type'] == 'Liquid', :]
data
```

```python
# Total up the sales for each Product Size and Scent for each Store
data = data.groupby(['Store Name', 'Size', 'Scent Name'])['Sale Value'].sum()
data = pd.DataFrame(data)
data = data.reset_index()
data
```

```python
# Rank each of the Product Size and Scent combinations for each Store
data['Rank of Product & Scent by Store'] = data.groupby(['Store Name'])['Sale Value'].rank(method='dense', ascending=False)
data
```

```python
# Only leave the top 10 based on total sales value calculated above
data = data.loc[data['Rank of Product & Scent by Store'] <= 10, :]
data['Rank of Product & Scent by Store'] = data['Rank of Product & Scent by Store'].astype(int)
data.head(10)
```

```python
# Round the Sales Values to the nearest 10 value
data['Sale Value'] = round(data['Sale Value'], 0)
data['Sale Value'] = data['Sale Value'].astype(int)
data.head(10)
```

```python
msg = "Which store would you like to view?"
title = "Store Selector"
choices=['Chelsea', 'Notting Hill', 'Dulwich', 'Lewisham', 'Shoreditch', 'Wimbledon', 'Cancel']

while True:
    result1 = g.buttonbox(msg, title, choices=choices)
    if result1 == 'Chelsea':
        data1 = data.loc[data['Store Name'] == 'Chelsea', :].sort_values(by=['Rank of Product & Scent by Store'])
        result = g.ynbox(data1, 'Continuation Message')
        if result ==  True:
            g.buttonbox(msg, title, choices=choices)
        else:
            sys.exit()
    elif result1 == 'Notting Hill':
        data2 = data.loc[data['Store Name'] == 'Notting Hill', :].sort_values(by=['Rank of Product & Scent by Store'])
        g.ccbox(data2, 'Results')
        if g.ccbox('Are you sure you would like to continue?', 'Confirmation Window'):
            g.buttonbox(msg, title, choices=choices)
        else:
            sys.exit()
    elif result1 == 'Dulwich':
        data3 = data.loc[data['Store Name'] == 'Dulwich', :].sort_values(by=['Rank of Product & Scent by Store'])
        g.ccbox(data3, 'Results')
        if g.ccbox('Are you sure you would like to continue?', 'Confirmation Window'):
            g.buttonbox(msg, title, choices=choices)
        else:
            sys.exit()
    elif result1 == 'Lewisham':
        data4 = data.loc[data['Store Name'] == 'Lewisham', :].sort_values(by=['Rank of Product & Scent by Store'])
        g.ccbox(data4, 'Results')
        if g.ccbox('Are you sure you would like to continue?', 'Confirmation Window'):
            g.buttonbox(msg, title, choices=choices)
        else:
            sys.exit()
    elif result1 == 'Shoreditch':
        data5 = data.loc[data['Store Name'] == 'Shoreditch', :].sort_values(by=['Rank of Product & Scent by Store'])
        g.ccbox(data5, 'Results')
        if g.ccbox('Are you sure you would like to continue?', 'Confirmation Window'):
            g.buttonbox(msg, title, choices=choices)
        else:
            sys.exit()
    elif result1 == 'Wimbledon':
        data6 = data.loc[data['Store Name'] == 'Wimbledon', :].sort_values(by=['Rank of Product & Scent by Store'])
        g.ccbox(data6, 'Results')
        if g.ccbox('Are you sure you would like to continue?', 'Confirmation Window'):
            g.buttonbox(msg, title, choices=choices)
        else:
            sys.exit()
    else:
        sys.exit()
```
