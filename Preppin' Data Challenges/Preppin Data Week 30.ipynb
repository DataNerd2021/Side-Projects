```python
import pandas as pd
import numpy as np
```

```python
# Input the Top 3 Sales People per Store for both regions
topsales_east = pd.read_csv("C:\\Users\\Chase\\Downloads\\Preppin' Summer 2022 - Top 3 Sales People per Store (east).csv")
topsales_west = pd.read_csv("C:\\Users\\Chase\\Downloads\\Preppin' Summer 2022 - Top 3 Sales People per Store (west).csv")
```

```python
# Combine these files and create Region field
topsales = pd.concat([topsales_west, topsales_east])
topsales['Region'] = np.where((topsales['Store'] == 1) | (topsales['Store'] == 3) | (topsales['Store'] == 5), 'East', 'West')
topsales
```

```python
# Input the Store Lookup file to provide the names of the Store instead of the ID number
stores = pd.read_csv("C:\\Users\\Chase\\Downloads\\Preppin' Summer 2022 - Store Lookup.csv")
df = topsales.merge(stores, how='inner', left_on='Store', right_on='StoreID')

# Remove any duplicate fields you have in the data set so far
df = df.drop(columns=['StoreID', 'Store', 'Region_y'])
df
```

```python
# Input the Week 27 Input file
data = pd.read_csv("C:\\Users\\Chase\\Downloads\\Preppin' Summer 2022 - PD 2022 Wk 27 input.csv")

# Use Week 27 Input file to create Sales Values for each Store
data = data.groupby(['Store Name', 'Region'])['Sale Value'].sum()
data = pd.DataFrame(data)
data = data.reset_index()
data
```

```python
# Combine this data with the rest of the prepared data
df = df.merge(data, how='inner')
 # Use the data set you have created to determine the actual sales value (rather than percentage) for each sales person
df['Sales per Person'] = round(df['Sale Value'] * (df['Percent of Store Sales']/100), 2)

# Remove any remain duplicate fields and reorder columns
df = df.drop(columns = ['Region_x'])
df = df[['Sales per Person', 'Region', 'Store Name', 'Sales Person', 'Percent of Store Sales', 'Sale Value']]
df = df.sort_values(by=['Sale Value', 'Percent of Store Sales'], ascending=[False, True])
df
```
