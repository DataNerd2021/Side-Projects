### Step 1: Import libraries and data

```python
import pandas as pd 
import numpy as np

df = pd.read_csv("C:\\Users\\chase\\Downloads\\Preppin' Summer 2022 - PD 2022 Wk 27 Input.csv")
df
```

### Step 2: Convert any data types required (Not necessary)

```python
df.info()
```

### Step 3: Create a new row for each day that doesn't have a sale

```python
df = df.groupby('Sale Date')['Sale Value'].sum().reset_index()
df.index = pd.DatetimeIndex(df['Sale Date']).floor('D')
all_days = pd.date_range(df.index.min(), df.index.max(), freq='D')
df = df.reindex(all_days)
df = df.reset_index()
df['Sale Date'] = df['index']
df = df.drop('index', axis=1)
df['Sale Value'] = df['Sale Value'].fillna(0)
df
```

### Step 4: Remove any date record where a sale occurred.

```python
condition = df[df['Sale Value'] > 0.0].index
df = df.drop(condition, inplace=False)
df = pd.DataFrame(df)
df
```

### Step 5: Create a column for Day of the Week

```python
df['Day of the Week'] = df['Sale Date'].dt.day_name()
df
```

### Step 6: For each day of the week, count the number of dates where there were no sales

```python
df = df.groupby('Day of the Week')['Day of the Week'].count()
df = pd.DataFrame(df)
df = df.rename(columns={'Day of the Week': 'Number of Days'})
df = df.reset_index()
df['Day of the Week'] = pd.Categorical(df['Day of the Week'], ["Wednesday", 'Tuesday', 'Sunday', 'Saturday', 'Monday', 'Thursday', 'Friday'])
df = df.sort_values('Day of the Week')
df
```
